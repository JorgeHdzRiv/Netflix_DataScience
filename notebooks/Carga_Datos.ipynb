# Los Datos

## **netflix_dataset.csv**
| Columna       | Descripción |
|---------------|-------------|
| show_id       | Identificador único del show |
| type          | Tipo de contenido (Película o Serie) |
| title         | Título del show |
| director      | Director del show |
| cast          | Reparto del show |
| country       | País de origen del show |
| date_added    | Fecha en que el show fue añadido a Netflix |
| release_year  | Año de lanzamiento del show |
| rating        | Clasificación del show (por edad) |
| duration      | Duración del show (en minutos o número de temporadas) |
| listed_in     | Categorías en las que el show está listado |
| description   | Descripción breve del show |



```python
# Importar librerias necesarias
import pandas as pd
```
